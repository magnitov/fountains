```python
plumes_fontanka = pd.read_csv('fountains_fontanka_merged.bed', sep = '\t', header = None)
plumes_fontanka.columns = ['chrom', 'start', 'end']
plumes_fontanka.head()
```

```python
def get_pileup(clr, expected_cooler, annotation, flank = 500_000, view_df = mm10_view_df):
    stack = cooltools.pileup(clr, annotation, view_df = mm10_view_df, expected_df = expected_cooler, 
                             flank = flank, nproc = 32)
    mtx = np.nanmean(stack, axis=2)
    return(mtx)
```
