```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
chromhmm_merge = pd.read_csv('chromhmm_merged.csv')
chromhmm = pd.DataFrame([chromhmm_merge['V1'].values, chromhmm_merge['V2'].values]).transpose()   
chromhmm[1] = [float(x) for x in chromhmm[1].values]
chromhmm
```

```python
plt.figure(figsize = (0.7, 6))
plt.title('Plumes chromHMM states')
sns.heatmap(data = np.round(np.log2(chromhmm.drop([0], axis = 1)), 1), xticklabels = 0, yticklabels = chromhmm[0].values, 
            annot = True, cmap = 'coolwarm', center = 0, vmin = -1.5, vmax = 1.5,
            cbar = 1, cbar_kws = {'shrink': 0.6, 'label': 'log2(fold enrichment)'})
plt.show()
```
